### Process for clean Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('fipkart_mobiles.csv')
dff=pd.read_csv('flipkart_mobile_scraped.csv')
```

```python
df1=pd.concat([df,dff],ignore_index=True)
```

```python
df2.columns = df2.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
for col in ['original_price', 'selling_price', 'rating']:
    df2[col] = df2[col].fillna(
        df2.groupby(['brand', 'model'])[col].transform('mean')
    )
```

```python
df2.isna().sum()
```

```python
def fill_with_mode(series):
    mode_vals = series.mode().dropna()
    return series.fillna(mode_vals.iloc[0]) if not mode_vals.empty else series

for col in ['model', 'color', 'memory', 'storage']:
    df2[col] = df2.groupby('brand')[col].transform(fill_with_mode)
```
